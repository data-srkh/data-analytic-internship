```python
trial_stores = data[data['STORE_NBR'].isin([77, 86, 88])]
#trial_stores = trial_stores[(trial_stores['DATE'] >= '2023-01-01') & (trial_stores['DATE'] <= '2023-06-30')]
```

#Aggregating Monthly Metrics

```python
trial_stores['DATE'] = pd.to_datetime(trial_stores['DATE'])
trial_stores['MONTH'] = trial_stores['DATE'].dt.to_period('M')
trial_stores
```

```python
monthly_metrics = trial_stores.groupby(['STORE_NBR','MONTH']).agg(
    total_sales = ('TOT_SALES','sum'),
    total_customers = ('LYLTY_CARD_NBR','nunique'),
    total_trasnactions = ('TXN_ID','nunique')
).reset_index()
monthly_metrics['avg_transactions_per_customer'] = monthly_metrics['total_sales']/monthly_metrics['total_customers']
monthly_metrics
```
