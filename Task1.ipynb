#Finding and removing outliers

I used Z score method to detect outliers.


```python
z_scores = np.abs(zscore(purchase_data.select_dtypes(include=[np.number])))
threshold = 3
outliers = (z_scores > threshold).any(axis=1)
print("Outliers detected using Z-score:")
print(purchase_data[outliers])
```

```python
purchase_df = purchase_data[~outliers]
print("DataFrame after removing outliers:")
print(purchase_df)
```

```python
z_scores = np.abs(zscore(transaction_data.select_dtypes(include=[np.number])))
threshold = 3
outliers = (z_scores > threshold).any(axis=1)
print("Outliers detected using Z-score:")
print(transaction_data[outliers])
```

```python
transaction_df = transaction_data[~outliers]
print("DataFrame after removing outliers:")
print(transaction_df)
```

#Checking and correcting data formats

```python
purchase_df['LYLTY_CARD_NBR'] = pd.to_numeric(purchase_df['LYLTY_CARD_NBR'], errors='coerce')
```

```python
transaction_df['STORE_NBR'] = pd.to_numeric(transaction_df['STORE_NBR'], errors='coerce')
transaction_df['LYLTY_CARD_NBR'] = pd.to_numeric(transaction_df['LYLTY_CARD_NBR'], errors='coerce')
transaction_df['TXN_ID'] = pd.to_numeric(transaction_df['TXN_ID'], errors='coerce')
transaction_df['PROD_NBR'] = pd.to_numeric(transaction_df['PROD_NBR'], errors='coerce')
transaction_df['PROD_QTY'] = pd.to_numeric(transaction_df['PROD_QTY'], errors='coerce')
transaction_df['TOT_SALES'] = pd.to_numeric(transaction_df['TOT_SALES'], errors='coerce')
```

```python
transaction_df['DATE'] = pd.to_datetime(transaction_df['DATE'], format='%Y-%m-%d', errors='coerce')
```
